```python
import pandas as pd
```

```python
ml100k_interactions = pd.read_csv('/workspace/gregorio/gobrec/experiments/datasets/ml-100k/u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
ml100k_interactions
```

```python
ml100k_items = pd.read_csv('/workspace/gregorio/gobrec/experiments/datasets/ml-100k/u.item', sep='|', names=['item_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'], encoding='latin-1')
ml100k_items
```

```python
genres = ['unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
ml100k_items = ml100k_items[['item_id'] + genres]
ml100k_items
```

```python
ml100k_interactions_items = ml100k_interactions.merge(ml100k_items, on='item_id', how='left')
ml100k_interactions_items
```

```python
ml100k_interactions_items['genres_sum'] = ml100k_interactions_items[genres].sum(axis=1)
ml100k_interactions_items
```

```python
ml100k_cumsum = ml100k_interactions_items.groupby('user_id')[['genres_sum'] + genres].cumsum()
contexts = ml100k_cumsum[genres].to_numpy() / ml100k_cumsum['genres_sum'].to_numpy()[:, None]
contexts
```

```python
ml100k_interactions_items[ml100k_interactions['user_id'] == 1][genres]
```

```python
pd.DataFrame(contexts[ml100k_interactions[ml100k_interactions['user_id'] == 1].index.to_numpy()], columns=genres)
```

```python
# Precisa arrumar problema que os contextos incluem o item atual.
import numpy as np

for user_id in ml100k_interactions['user_id'].unique():
    user_indices = ml100k_interactions[ml100k_interactions['user_id'] == user_id].index.to_numpy()
    user_contexts = contexts[user_indices]
    user_contexts_shifted = np.roll(user_contexts, shift=1, axis=0)
    user_contexts_shifted[0] = 0  # Primeiro contexto é zero
    contexts[user_indices] = user_contexts_shifted
```
